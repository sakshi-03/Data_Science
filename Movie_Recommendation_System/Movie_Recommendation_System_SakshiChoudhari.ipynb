```python
mean_vote = mdf['vote_average'].mean()
print(mean_vote)
```

```python
mini_vote=mdf['vote_count'].quantile(0.90)
print(mini_vote)
```

```python
q_movies=mdf.copy().loc[mdf['vote_count']>=mini_vote]
q_movies.shape
```

```python
def weighted_rating(x,mini_vote=mini_vote, mean_vote=mean_vote):
    voters=x['vote_count']
    avg_vote=x['vote_average']
    return(voters/(voters+mini_vote)*avg_vote)+(mini_vote/(mini_vote+voters)*mean_vote)
```

```python
q_movies['score']=q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies=q_movies.sort_values('score',ascending=False)
```
